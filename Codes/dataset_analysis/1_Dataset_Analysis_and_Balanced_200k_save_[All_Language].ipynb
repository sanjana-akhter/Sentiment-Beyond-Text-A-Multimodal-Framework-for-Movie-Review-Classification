# **Language Checker**

```python
# =========================================
#  Detect Review Text Languages (20k sample)
# =========================================
!pip install langdetect langcodes > /dev/null

import pandas as pd
from langdetect import detect, DetectorFactory
from langcodes import Language

DetectorFactory.seed = 0

# --- safe detection ---
def detect_lang_safe(text: str) -> str:
    try:
        return detect(str(text))
    except Exception:
        return "und"

# --- code -> human name  ---
def lang_display_name(code: str) -> str:

    if not code or str(code).lower() in {"und", "unknown", "nan"}:
        return "Undetermined"
    try:
        return Language.get(str(code)).display_name("en")  # e.g., 'Portuguese (Brazil)'
    except Exception:
        # if something odd like 'zh_cn', normalize and retry
        norm = str(code).replace("_", "-")
        try:
            return Language.get(norm).display_name("en")
        except Exception:
            return str(code).upper()

# ------- sample 20k and detect -------
SAMPLE_N = 20000
cols_needed = ['id', 'title', 'reviewText', 'originalLanguage']
sample_reviews = (
    merged_df[cols_needed]
    .dropna(subset=['reviewText'])
    .sample(SAMPLE_N, random_state=47)
    .copy()
)

sample_reviews['reviewLang_code'] = sample_reviews['reviewText'].apply(detect_lang_safe)
sample_reviews['reviewLang_name'] = sample_reviews['reviewLang_code'].apply(lang_display_name)
```

```python
# =========================================
# üìä 1) Distribution: Count & Percentage
# =========================================
lang_counts = sample_reviews['reviewLang_name'].value_counts(dropna=False)
lang_percent = (lang_counts / lang_counts.sum() * 100)

lang_summary = (
    pd.DataFrame({
        'Review Text Language': lang_counts.index,
        'Count': lang_counts.values,
        'Percentage': lang_percent.values.round(2)
    })
    .sort_values('Count', ascending=False)
    .reset_index(drop=True)
)

print("Language distribution in 20k sample (by review text language):")
display(lang_summary)
```

```python
# =========================================
#  Top-3 movie originalLanguage per review-text language
# =========================================
# Clean up missing originalLanguage
sample_reviews = sample_reviews.copy()
sample_reviews['originalLanguage'] = sample_reviews['originalLanguage'].fillna('Unknown')

def top3_original_langs(group: pd.DataFrame) -> pd.Series:
    vc = group['originalLanguage'].astype(str).value_counts()
    top3 = vc.head(3)
    total = vc.sum()
    # include percentages too
    fmt = lambda lang, cnt: f"{lang}: {int(cnt)} ({cnt/total*100:.2f}%)"
    out = {
        'Top 1': fmt(top3.index[0], top3.iloc[0]) if len(top3) > 0 else "",
        'Top 2': fmt(top3.index[1], top3.iloc[1]) if len(top3) > 1 else "",
        'Top 3': fmt(top3.index[2], top3.iloc[2]) if len(top3) > 2 else "",
    }
    return pd.Series(out)

top3_table = (
    sample_reviews
    .groupby('reviewLang_name')
    .apply(top3_original_langs, include_groups=False)  # <-- silences the warning
    .reset_index()
    .rename(columns={'reviewLang_name': 'Review Text Language'})
    .sort_values('Review Text Language')
    .reset_index(drop=True)
)

display(top3_table)
```

```python
# =========================================
#  3) Non-English review texts only
# =========================================
non_en = sample_reviews[sample_reviews['reviewLang_code'] != 'en'].copy()

non_en_counts = non_en['reviewLang_name'].value_counts()
non_en_percent = (non_en_counts / len(sample_reviews) * 100)

non_en_summary = (
    pd.DataFrame({
        'Non-English Review Text Language': non_en_counts.index,
        'Count': non_en_counts.values,
        'Percentage_of_total_sample': non_en_percent.values.round(2)
    })
    .sort_values('Count', ascending=False)
    .reset_index(drop=True)
)

print("\nNon-English review texts (counts & % of the 20k sample):")
display(non_en_summary)
```

```python
# ============================================================
#  Normalize movie originalLanguage into target buckets
#            (English + Spanish/Hindi/Japanese/Chinese/Portuguese/French/German)
# ============================================================
import re
import pandas as pd

def norm_movie_lang(s: str) -> str:
    s0 = (str(s) or "").strip()
    s = s0.lower()

    # english variants
    if 'english' in s or re.search(r'\ben(-|_|$)|\beng\b', s):
        return 'English'

    # spanish
    if any(k in s for k in ['spanish','espa√±ol','castilian','castellano']) or re.search(r'\bes(-|_|$)', s):
        return 'Spanish'

    # hindi (Devanagari or code)
    if 'hindi' in s or re.search(r'\bhi(-|_|$)', s):
        return 'Hindi'

    # japanese
    if any(k in s for k in ['japanese','Êó•Êú¨Ë™û']) or re.search(r'\bja(-|_|$)', s):
        return 'Japanese'

    # chinese (lump variants)
    if any(k in s for k in ['chinese','‰∏≠Êñá','mandarin','cantonese','Á≤§ËØ≠']) or re.search(r'\bzh(-|_|$)|\byue\b', s):
        return 'Chinese'

    # portuguese
    if any(k in s for k in ['portuguese','portugu√™s','portugues']) or re.search(r'\bpt(-|_|$)', s):
        return 'Portuguese'

    # french
    if any(k in s for k in ['french','fran√ßais','francais']) or re.search(r'\bfr(-|_|$)', s):
        return 'French'

    # german
    if any(k in s for k in ['german','deutsch']) or re.search(r'\bde(-|_|$)', s):
        return 'German'

    # fallback: keep original (useful for diagnostics, but won't be used in "other 30k")
    return s0 if s0 else 'Unknown'

filtered_df = filtered_df.copy()
filtered_df['movieLangNorm'] = filtered_df['originalLanguage'].apply(norm_movie_lang)

print("Movie language buckets (top 15):")
print(filtered_df['movieLangNorm'].value_counts().head(15))
```

```python
# ============================================================
#     Create Dataset A with natural availability  (WITH ENGLISH TITLE CAP)
#   - Uses all available NEGATIVE from the 7-langs pool (<= 15k)
#   - Matches POSITIVE to that same count (balanced within "other" block)
#   - Fills the rest with English films to hit 200k and 100k/100k
#   - For ENGLISH only: ‚â§ 16 reviews per movie title (total across POS+NEG)
# ============================================================

RNG_OTHER = 47
RNG_EN_A  = 57
RNG_EN_B  = 97
CAP_PER_TITLE_EN = 16 # <= 16 reviews per English title

SAVE_PATH = '/content/drive/MyDrive/THESIS 47/Saved Datasets/Dataset_A_200k_balanced_english_cap5.csv'

# ---------------------------
# 1) "Other" (non-English) block
# ---------------------------
neg_avail_other = (other_films['scoreSentiment'] == 'NEGATIVE').sum()
target_other_neg = min(15_000, neg_avail_other)

other_neg_blk = other_films[other_films['scoreSentiment'] == 'NEGATIVE'].sample(
    n=target_other_neg, random_state=RNG_OTHER
)
other_pos_blk = other_films[other_films['scoreSentiment'] == 'POSITIVE'].sample(
    n=target_other_neg, random_state=RNG_OTHER
)
other_block = pd.concat([other_pos_blk, other_neg_blk], axis=0)

print(f"[Other block] POS={len(other_pos_blk)}, NEG={len(other_neg_blk)}, TOTAL={len(other_block)}")

# ---------------------------
# 2) English quotas to complete 200k and 100k/100k
# ---------------------------
NEED_POS_ALL = 100_000
NEED_NEG_ALL = 100_000

rem_pos_eng = NEED_POS_ALL - (other_block['scoreSentiment'] == 'POSITIVE').sum()
rem_neg_eng = NEED_NEG_ALL - (other_block['scoreSentiment'] == 'NEGATIVE').sum()

print(f"[English quota] POS={rem_pos_eng}, NEG={rem_neg_eng}")

# Sanity: check English availability (before capping)
avail_pos_eng = (english_films['scoreSentiment'] == 'POSITIVE').sum()
avail_neg_eng = (english_films['scoreSentiment'] == 'NEGATIVE').sum()
assert avail_pos_eng >= rem_pos_eng, f"Not enough English POS: need {rem_pos_eng}, have {avail_pos_eng}"
assert avail_neg_eng >= rem_neg_eng, f"Not enough English NEG: need {rem_neg_eng}, have {avail_neg_eng}"

# ---------------------------
# 2a) Helper to enforce cap per English title (across POS+NEG)
# ---------------------------
def capped_take_by_title(df_in, n_needed, selected_counts, cap, title_col='title', rng=57):
    """
    Select up to n_needed rows from df_in while respecting a per-title cap
    tracked in selected_counts (shared across POS/NEG). Deterministic via rng.
    """
    if n_needed <= 0:
        return df_in.iloc[0:0], selected_counts

    df_shuf = df_in.sample(frac=1.0, random_state=rng)
    picks = []
    for idx, row in df_shuf.iterrows():
        t = row[title_col]
        c = selected_counts.get(t, 0)
        if c < cap:
            picks.append(idx)
            selected_counts[t] = c + 1
            if len(picks) == n_needed:
                break
    return df_in.loc[picks], selected_counts

# Pools
required_cols = {'title', 'scoreSentiment'}
missing = required_cols - set(english_films.columns)
if missing:
    raise KeyError(f"english_films is missing required columns: {missing}")

eng_pos_pool = english_films[english_films['scoreSentiment'] == 'POSITIVE']
eng_neg_pool = english_films[english_films['scoreSentiment'] == 'NEGATIVE']

# Shared counter of how many selected per English title (POS+NEG combined)
selected_per_title = {}

# Pick POS then NEG (switch order if you want to prioritize NEG)
eng_pos_sel, selected_per_title = capped_take_by_title(
    eng_pos_pool, rem_pos_eng, selected_per_title, CAP_PER_TITLE_EN,
    title_col='title', rng=RNG_EN_A
)
eng_neg_sel, selected_per_title = capped_take_by_title(
    eng_neg_pool, rem_neg_eng, selected_per_title, CAP_PER_TITLE_EN,
    title_col='title', rng=RNG_EN_A + 1
)

# Verify quotas met under cap; fail with a helpful message if not
if len(eng_pos_sel) < rem_pos_eng or len(eng_neg_sel) < rem_neg_eng:
    short_pos = rem_pos_eng - len(eng_pos_sel)
    short_neg = rem_neg_eng - len(eng_neg_sel)
    cap_hits = sum(1 for v in selected_per_title.values() if v >= CAP_PER_TITLE_EN)
    raise RuntimeError(
        "‚ùå Could not satisfy English quotas under the ‚â§5 reviews/title cap.\n"
        f"   Needed POS: {rem_pos_eng}, Selected: {len(eng_pos_sel)}, Short: {max(short_pos,0)}\n"
        f"   Needed NEG: {rem_neg_eng}, Selected: {len(eng_neg_sel)}, Short: {max(short_neg,0)}\n"
        f"   Titles at cap ({CAP_PER_TITLE_EN}): {cap_hits}\n"
        "   Consider increasing CAP_PER_TITLE_EN or adjusting quotas."
    )

english_block = pd.concat([eng_pos_sel, eng_neg_sel], axis=0)

# Double-check cap on ENGLISH only
over_cap = english_block['title'].value_counts()
over_cap = over_cap[over_cap > CAP_PER_TITLE_EN]
assert over_cap.empty, f"Found English titles over cap: \n{over_cap.head()}"

print(f"[English block] POS={len(eng_pos_sel)}, NEG={len(eng_neg_sel)}, TOTAL={len(english_block)}")
print(f"[English block] Titles used: {english_block['title'].nunique()} (cap per title = {CAP_PER_TITLE_EN})")

# ---------------------------
# 3) Final Dataset A (shuffle + save)
# ---------------------------
dataset_a = (
    pd.concat([english_block, other_block], axis=0)
    .sample(frac=1, random_state=RNG_EN_B)
    .reset_index(drop=True)
)

print("‚úÖ Dataset A shape:", dataset_a.shape)
print("‚úÖ Dataset A sentiment distribution:\n", dataset_a['scoreSentiment'].value_counts())
if 'movieLangNorm' in dataset_a.columns:
    print("‚úÖ Dataset A movieLangNorm distribution (top 10):\n", dataset_a['movieLangNorm'].value_counts().head(10))

# Save for later reuse
dataset_a.to_csv(SAVE_PATH, index=False)
print(f"üíæ Saved Dataset A to: {SAVE_PATH}")

# Optional quick cap check inside final dataset (English only)
if 'movieLangNorm' in dataset_a.columns:
    is_eng = dataset_a['movieLangNorm'].str.lower().eq('english')
else:
    # fallback heuristic if you don't have movieLangNorm
    is_eng = dataset_a.get('originalLanguage', pd.Series(index=dataset_a.index, dtype=str))\
                      .str.contains('English', case=False, na=False)

cap_check = (dataset_a[is_eng]
             .groupby('title').size()
             .sort_values(ascending=False)
             .head(10))
print("\n[Cap check] Top English titles by selected review count (should be ‚â§ 16):")
print(cap_check)
```
